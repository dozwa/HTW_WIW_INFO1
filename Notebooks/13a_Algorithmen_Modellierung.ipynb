## 1. Was ist ein Algorithmus?

### Einführung und Motivation

In den bisherigen Notebooks haben Sie gelernt, wie man Variablen verwendet, Funktionen definiert und mit Schleifen arbeitet. Sie haben damit bereits viele **Programme** geschrieben. Doch was unterscheidet ein Programm von einem **Algorithmus**?

Ein **Algorithmus** ist eine präzise, endliche Folge von Anweisungen zur Lösung eines klar definierten Problems. Algorithmen sind sprachunabhängig – derselbe Algorithmus kann in Python, Java oder einer anderen Programmiersprache implementiert werden. Ein **Programm** hingegen ist die konkrete Umsetzung eines Algorithmus in einer bestimmten Programmiersprache.

Algorithmen sind das konzeptionelle Fundament der Informatik. Sie beschreiben **wie** ein Problem gelöst wird, unabhängig davon, in welcher Sprache oder auf welchem Computer die Lösung später ausgeführt wird.

### Eigenschaften eines Algorithmus

Ein korrekter Algorithmus muss folgende fünf Eigenschaften erfüllen:

1. **Eindeutigkeit**: Jeder Schritt ist präzise definiert und lässt keine Interpretationsspielräume
2. **Endlichkeit der Beschreibung**: Der Algorithmus muss in endlich vielen Zeilen beschrieben werden können
3. **Terminiertheit**: Der Algorithmus muss nach endlich vielen Schritten enden
4. **Determinismus**: Bei gleicher Eingabe liefert der Algorithmus stets dasselbe Ergebnis
5. **Effektivität**: Jeder Schritt muss tatsächlich ausführbar sein

Diese formalen Eigenschaften stellen sicher, dass Algorithmen zuverlässig und reproduzierbar funktionieren.

### Beispiele für Algorithmen aus dem Alltag

Algorithmen begegnen uns nicht nur in der Informatik, sondern auch im Alltag:

- **Kochrezept**: Eine Schritt-für-Schritt-Anleitung zum Zubereiten eines Gerichts
- **Wegbeschreibung**: Anweisungen, um von Punkt A nach Punkt B zu gelangen
- **Montageanleitung**: Schritte zum Zusammenbauen eines Möbelstücks

Diese Beispiele zeigen: Ein Algorithmus ist immer eine **Handlungsanweisung** zur systematischen Lösung eines Problems.

### Beispiel 1: Ein einfacher Algorithmus in Python

Betrachten Sie folgenden Algorithmus, der zwei Zahlen addiert – der einfachste denkbare Algorithmus:

````python
def addiere(a, b):
    """Addiert zwei Zahlen und gibt das Ergebnis zurück."""
    ergebnis = a + b
    return ergebnis

# Algorithmus anwenden
summe = addiere(5, 3)
print(f"5 + 3 = {summe}")
````

**Erwartete Ausgabe**:
```
5 + 3 = 8
```

**Erklärung**: Dieser Algorithmus ist eindeutig (nur eine Addition), deterministisch (gleiche Eingaben → gleiches Ergebnis) und terminiert sofort nach einer Operation. Er erfüllt alle fünf Eigenschaften eines korrekten Algorithmus.

### Beispiel 2: Algorithmus mit Schleife

Ein etwas komplexerer Algorithmus berechnet die Summe aller Zahlen von 1 bis n:

````python
def summe_bis_n(n):
    """Berechnet die Summe aller Zahlen von 1 bis n."""
    summe = 0
    for i in range(1, n + 1):
        summe = summe + i
    return summe

# Algorithmus anwenden
ergebnis = summe_bis_n(5)
print(f"Summe von 1 bis 5: {ergebnis}")
````

**Erwartete Ausgabe**:
```
Summe von 1 bis 5: 15
```

**Erklärung**: Der Algorithmus durchläuft alle Zahlen von 1 bis n und addiert sie schrittweise zur Variable `summe`. Am Ende wird die Gesamtsumme zurückgegeben. Dieser Algorithmus terminiert nach genau n Schleifendurchläufen.

## 2. Warum Algorithmen modellieren?

### Das Problem: Von der Idee zum Code

Stellen Sie sich vor, Sie sollen ein komplexes Problem lösen. Wo fangen Sie an? Direkt Code zu schreiben ist verlockend, führt aber oft zu:

- **Unstrukturiertem Code**: Ohne Plan wird der Code schnell unübersichtlich
- **Logikfehlern**: Fehler in der Grundlogik sind im fertigen Code schwer zu finden
- **Kommunikationsproblemen**: Wie erklären Sie Ihre Lösung anderen?
- **Zeitverlust**: Nachträgliche Änderungen der Grundstruktur sind aufwendig

Die Lösung: **Erst planen, dann programmieren!**

### Vorteile der Algorithmus-Modellierung

Durch das Modellieren von Algorithmen vor dem Programmieren erreichen Sie:

1. **Klare Struktur**: Sie verstehen das Problem vollständig, bevor Sie Code schreiben
2. **Frühe Fehlererkennung**: Logikfehler werden sichtbar, bevor Sie programmieren
3. **Bessere Kommunikation**: Diagramme und Pseudocode sind für alle verständlich – auch für Nicht-Programmierer
4. **Sprachunabhängigkeit**: Die Modellierung funktioniert unabhängig von der Programmiersprache
5. **Dokumentation**: Diagramme dienen als Dokumentation für spätere Wartung

In der professionellen Softwareentwicklung ist die Modellierungsphase ein fester Bestandteil des Entwicklungsprozesses.

## 3. Pseudocode

### Was ist Pseudocode?

**Pseudocode** ist eine informelle, aber strukturierte Beschreibung eines Algorithmus in natürlicher Sprache. Er verwendet Schlüsselwörter, die an Programmiersprachen erinnern, ist aber nicht an eine bestimmte Syntax gebunden.

Pseudocode dient als **Zwischenschritt** zwischen der Idee und dem fertigen Programm. Er ermöglicht es, die Logik eines Algorithmus zu formulieren, ohne sich um syntaktische Details einer Programmiersprache kümmern zu müssen.

### Konventionen für Pseudocode

Es gibt keine verbindliche Norm für Pseudocode, aber folgende Konventionen haben sich bewährt:

| Schlüsselwort | Bedeutung | Python-Äquivalent |
|---------------|-----------|-------------------|
| `ALGORITHMUS` | Funktionsdefinition | `def` |
| `EINGABE` | Parameter | Parameter in Klammern |
| `AUSGABE` | Rückgabewert | `return` |
| `←` | Zuweisung | `=` |
| `WENN...DANN` | Bedingung | `if` |
| `SONST` | Alternative | `else` |
| `SOLANGE` | While-Schleife | `while` |
| `FÜR...VON...BIS` | For-Schleife | `for...in range()` |
| `RÜCKGABE` | Return | `return` |
| `MOD` | Modulo | `%` |

### Beispiel 3: Addition in Pseudocode

Beginnen wir mit dem einfachsten Algorithmus – der Addition zweier Zahlen:

```
ALGORITHMUS Addiere(a, b)
    EINGABE: Zwei Zahlen a und b
    AUSGABE: Die Summe von a und b
    
    ergebnis ← a + b
    RÜCKGABE ergebnis
ENDE
```

Dieser Pseudocode beschreibt klar:
- Was der Algorithmus erhält (Eingabe)
- Was er zurückgibt (Ausgabe)
- Wie er das Ergebnis berechnet (Schritte)

### Übersetzung in Python

Die Übersetzung von Pseudocode nach Python ist systematisch:

````python
# Pseudocode → Python
# ALGORITHMUS Addiere(a, b)  →  def addiere(a, b):
# ergebnis ← a + b          →  ergebnis = a + b
# RÜCKGABE ergebnis         →  return ergebnis

def addiere(a, b):
    """Addiert zwei Zahlen."""
    ergebnis = a + b
    return ergebnis

# Test
print(f"7 + 4 = {addiere(7, 4)}")
````

**Erwartete Ausgabe**:
```
7 + 4 = 11
```

### Beispiel 4: Summe von 1 bis n in Pseudocode

Ein Algorithmus mit Schleife:

```
ALGORITHMUS SummeBisN(n)
    EINGABE: Eine positive ganze Zahl n
    AUSGABE: Die Summe aller Zahlen von 1 bis n
    
    summe ← 0
    FÜR i VON 1 BIS n:
        summe ← summe + i
    ENDE FÜR
    RÜCKGABE summe

Beachten Sie die Einrückung innerhalb der Schleife – sie zeigt die Struktur des Algorithmus.

**Erwartete Ausgabe**:
```

### Beispiel 5: GGT (Euklidischer Algorithmus)

Der **größte gemeinsame Teiler (GGT)** zweier Zahlen ist die größte Zahl, die beide ohne Rest teilt. Der **Euklidische Algorithmus** ist ein eleganter Weg, den GGT zu berechnen:

```
ALGORITHMUS GGT(a, b)
    EINGABE: Zwei positive ganze Zahlen a und b
    AUSGABE: Der größte gemeinsame Teiler von a und b
    
    SOLANGE b ≠ 0:
        rest ← a MOD b
        a ← b
        b ← rest
    ENDE SOLANGE
    
    RÜCKGABE a
ENDE
```

**Idee**: Der GGT von a und b ist derselbe wie der GGT von b und (a mod b). Wiederhole dies, bis b=0 ist.

````python
# Übersetzung in Python
def ggt(a, b):
    """Berechnet den größten gemeinsamen Teiler von a und b."""
    while b != 0:
        rest = a % b
        a = b
        b = rest
    return a

# Tests
print(f"GGT(48, 18) = {ggt(48, 18)}")
print(f"GGT(17, 5) = {ggt(17, 5)}")
````

**Erwartete Ausgabe**:
```
GGT(48, 18) = 6
GGT(17, 5) = 1
```

**Erklärung für GGT(48, 18)**:
- Durchlauf 1: rest = 48 % 18 = 12, a = 18, b = 12
- Durchlauf 2: rest = 18 % 12 = 6, a = 12, b = 6
- Durchlauf 3: rest = 12 % 6 = 0, a = 6, b = 0
- b = 0, also Ende → Ergebnis: 6

## 4. UML-Ablaufdiagramme

### Was sind Ablaufdiagramme?

**Ablaufdiagramme** (auch Flussdiagramme genannt) sind grafische Darstellungen von Algorithmen. Sie verwenden standardisierte Symbole, um den Ablauf eines Programms zu visualisieren.

Die **UML** (Unified Modeling Language) definiert eine standardisierte Notation für solche Diagramme. UML-Ablaufdiagramme werden in der Softwareentwicklung weltweit verwendet und sind Teil vieler Entwicklungsmethoden.

**Vorteile von Ablaufdiagrammen**:
- Komplexe Verzweigungen werden sofort sichtbar
- Auch für Nicht-Programmierer verständlich
- Fehler in der Logik fallen schnell auf
- Dienen als Dokumentation

### Die UML-Symbole

Für Ablaufdiagramme verwenden wir folgende Grundsymbole:

| Symbol | Form | Bedeutung | Beispiel |
|--------|------|-----------|----------|
| **Start/Ende** | Oval (abgerundet) | Beginn oder Ende des Algorithmus | "Start", "Ende" |
| **Anweisung** | Rechteck | Eine Aktion oder Berechnung | "summe = summe + i" |
| **Entscheidung** | Raute | Eine Bedingung mit Ja/Nein | "i < n?" |
| **Ein-/Ausgabe** | Parallelogramm | Eingabe oder Ausgabe von Daten | "Eingabe: n" |
| **Pfeil** | Linie mit Pfeilspitze | Verbindung, zeigt Ablaufrichtung | → |

### Mermaid-Syntax für Ablaufdiagramme

Die grundlegende Syntax für ein Mermaid-Ablaufdiagramm:

```
flowchart TD
    A([Start]) --> B[Anweisung]
    B --> C{Bedingung?}
    C -->|Ja| D[Aktion wenn Ja]
    C -->|Nein| E[Aktion wenn Nein]
    D --> F([Ende])
    E --> F
```

**Erklärung der Syntax:**
- `flowchart TD`: Erstellt ein Flussdiagramm von oben nach unten (Top-Down)
- `([Text])`: Ovales Symbol (Start/Ende)
- `[Text]`: Rechteck (Anweisung)
- `{Text?}`: Raute (Entscheidung)
- `-->`: Pfeil zum nächsten Element
- `-->|Label|`: Pfeil mit Beschriftung

### Beispiel 7: Summe von 1 bis n als Ablaufdiagramm

Ein Algorithmus mit Schleife:

Ablaufdiagramm: Summe von 1 bis n

**Erklärung**: Die Raute (Entscheidung) prüft die Schleifenbedingung. Bei "Ja" wird der Schleifenkörper ausgeführt und zur Bedingung zurückgekehrt. Bei "Nein" wird die Schleife verlassen.

````python
# Python-Implementation basierend auf dem Diagramm
def summe_bis_n_diagramm(n):
    summe = 0           # Initialisierung
    i = 1               # Schleifenvariable
    while i <= n:       # Bedingung (Raute)
        summe = summe + i   # Schleifenkörper
        i = i + 1           # Inkrement
    return summe        # Ausgabe

print(f"Summe von 1 bis 100: {summe_bis_n_diagramm(100)}")
````

**Erwartete Ausgabe**:
```
Summe von 1 bis 100: 5050
```

### Beispiel 8: GGT als Ablaufdiagramm

Der Euklidische Algorithmus als Ablaufdiagramm:

Ablaufdiagramm: GGT (Euklidischer Algorithmus)

**Erklärung**: Die Schleife läuft, solange b nicht 0 ist. In jedem Durchlauf werden die Werte verschoben: a bekommt den Wert von b, b bekommt den Rest der Division.

````python
# Python-Implementation basierend auf dem Diagramm
def ggt_diagramm(a, b):
    while b != 0:           # Bedingung (Raute)
        rest = a % b        # rest = a MOD b
        a = b               # a = b
        b = rest            # b = rest
    return a                # Ausgabe

print(f"GGT(56, 98) = {ggt_diagramm(56, 98)}")
````

**Erwartete Ausgabe**:
```
GGT(56, 98) = 14
```

## 5. Primzahlenprüfung: Vollständiges Beispiel

### Das Problem

Eine **Primzahl** ist eine natürliche Zahl größer als 1, die nur durch 1 und sich selbst teilbar ist. Beispiele: 2, 3, 5, 7, 11, 13, ...

**Aufgabe**: Entwickeln Sie einen Algorithmus, der prüft, ob eine gegebene Zahl n eine Primzahl ist.

**Strategie**:
1. Zahlen kleiner als 2 sind keine Primzahlen
2. Prüfe alle möglichen Teiler von 2 bis n-1
3. Wenn ein Teiler gefunden wird → keine Primzahl
4. Wenn kein Teiler gefunden wird → Primzahl

### Schritt 1: Pseudocode entwickeln

```
ALGORITHMUS IstPrimzahl(n)
    EINGABE: Eine positive ganze Zahl n
    AUSGABE: WAHR wenn n eine Primzahl ist, sonst FALSCH
    
    WENN n < 2 DANN
        RÜCKGABE FALSCH
    ENDE WENN
    
    FÜR teiler VON 2 BIS n-1:
        WENN n MOD teiler = 0 DANN
            RÜCKGABE FALSCH
        ENDE WENN
    ENDE FÜR
    
    RÜCKGABE WAHR
ENDE
```

**Erklärung**: Der Algorithmus hat zwei Abbruchbedingungen – einmal für Zahlen < 2 und einmal, wenn ein Teiler gefunden wird. Nur wenn die Schleife komplett durchläuft ohne Teiler zu finden, ist die Zahl prim.

### Schritt 2: Ablaufdiagramm erstellen

Ablaufdiagramm: Primzahlprüfung

**Erklärung**: Das Diagramm zeigt die verschachtelte Struktur: Eine Schleife (äußere Bedingung `teiler < n?`) mit einer inneren Bedingung (Teilbarkeitstest). Beachten Sie die drei möglichen Ausgänge (früher Abbruch bei n<2, Teiler gefunden, kein Teiler gefunden).

### Schritt 3: Python-Implementation

````python
def ist_primzahl(n):
    """Prüft, ob n eine Primzahl ist."""
    # Sonderfall: Zahlen < 2 sind keine Primzahlen
    if n < 2:
        return False
    
    # Prüfe alle möglichen Teiler von 2 bis n-1
    for teiler in range(2, n):
        if n % teiler == 0:
            return False    # Teiler gefunden → keine Primzahl
    
    return True    # Kein Teiler gefunden → Primzahl
````

````python
# Tests
testzahlen = [1, 2, 3, 4, 5, 10, 11, 17, 20, 23]

for zahl in testzahlen:
    ergebnis = ist_primzahl(zahl)
    status = "ist Primzahl" if ergebnis else "ist keine Primzahl"
    print(f"{zahl} {status}")
````

## 6. Vom Diagramm zum Code: Systematische Übersetzung

### Die Übersetzungsregeln

Um ein Ablaufdiagramm systematisch in Python-Code zu übersetzen, verwenden Sie folgende Zuordnungen:

| Diagramm-Element | Python-Konstrukt |
|------------------|------------------|
| Start + Eingabe | `def funktionsname(parameter):` |
| Rechteck (Anweisung) | Einfache Anweisung |
| Raute mit Rückkehr | `while bedingung:` |
| Raute ohne Rückkehr | `if bedingung:` |
| Raute mit Zähler | `for variable in range(...):` |
| Ausgabe + Ende | `return wert` |

### Beispiel 9: Schritt-für-Schritt-Übersetzung

Betrachten Sie dieses Ablaufdiagramm für die Berechnung von n! (Fakultät):

Ablaufdiagramm: Fakultät

**Übersetzung Schritt für Schritt:**

````python
# Schritt 1: Start + Eingabe → Funktionskopf
def fakultaet_diagramm(n):
    
    # Schritt 2: ergebnis = 1 → Initialisierung
    ergebnis = 1
    
    # Schritt 3: i = 1 → Schleifenvariable
    i = 1
    
    # Schritt 4: Raute mit Rückkehr → while-Schleife
    while i <= n:
        # Schritt 5: ergebnis = ergebnis * i
        ergebnis = ergebnis * i
        # Schritt 6: i = i + 1
        i = i + 1
    
    # Schritt 7: Ausgabe + Ende → return
    return ergebnis

# Test
print(f"5! = {fakultaet_diagramm(5)}")
print(f"7! = {fakultaet_diagramm(7)}")
````

**Erwartete Ausgabe**:

def finde_minimum(liste):

    minimum = liste[0]
    for element in liste:
        if element < minimum:
            minimum = element
    return minimum

```python
def durchschnitt(liste):
    """Berechnet den Durchschnitt einer Liste von Zahlen."""
    if len(liste) == 0:
        return 0
    
    summe = 0
    for zahl in liste:
        summe = summe + zahl
    
    durchschnitt = summe / len(liste)
    return durchschnitt

# Tests
print(f"Durchschnitt von [10, 20, 30]: {durchschnitt([10, 20, 30])}")
print(f"Durchschnitt von []: {durchschnitt([])}")
```

Durchschnitt von []: 0
