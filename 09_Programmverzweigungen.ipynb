## 1. Was sind Programmverzweigungen? рџ¤”

### Die Idee hinter Verzweigungen

Stellen Sie sich vor, Sie fahren mit dem Auto und kommen an eine Kreuzung. Je nachdem, wohin Sie mГ¶chten, biegen Sie links, rechts ab oder fahren geradeaus weiter. Genau so funktionieren **Programmverzweigungen**: Ihr Programm kommt an einen Punkt, an dem es eine Entscheidung treffen muss вЂ“ welcher Weg soll eingeschlagen werden?

In der Programmierung nennen wir solche Entscheidungspunkte **bedingte Anweisungen** oder **Programmverzweigungen**. Sie ermГ¶glichen es, dass bestimmte Teile Ihres Codes nur dann ausgefГјhrt werden, wenn eine bestimmte **Bedingung** erfГјllt ist.

Ein einfaches Beispiel aus dem Alltag: "**Wenn** es regnet, **dann** nehme ich einen Regenschirm mit." Das "wenn" ist die Bedingung (es regnet), und das "dann" beschreibt die Aktion, die bei erfГјllter Bedingung ausgefГјhrt wird (Regenschirm mitnehmen).

In der Programmierung funktioniert das genauso: **Wenn** eine Bedingung wahr ist (z.B. eine Zahl ist grГ¶Гџer als 100), **dann** wird ein bestimmter Codeblock ausgefГјhrt (z.B. eine Warnung ausgeben).

### Warum brauchen wir Verzweigungen?

Ohne Programmverzweigungen wГ¤ren unsere Programme extrem limitiert. Sie wГјrden immer genau dasselbe tun, egal welche Eingaben Sie machen oder welche Situationen auftreten. Mit Verzweigungen kГ¶nnen wir:

- **Auf Benutzereingaben reagieren**: Wenn der Benutzer "ja" eingibt, passiert etwas anderes als bei "nein"
- **Fehler abfangen**: Wenn eine ungГјltige Eingabe gemacht wird, kГ¶nnen wir eine Fehlermeldung ausgeben
- **Komplexe Logik umsetzen**: Produktionsanlagen, Alarmsysteme, Berechnungen вЂ“ all das benГ¶tigt Entscheidungslogik
- **Programme intelligent machen**: Ihr Code kann sich an verschiedene Situationen anpassen

### Die wichtigsten Konzepte

Bevor wir in die Details einsteigen, hier die wichtigsten Begriffe:

- **Bedingung**: Ein Ausdruck, der entweder wahr (`True`) oder falsch (`False`) ist. Zum Beispiel: `x > 10`
- **Anweisungsblock**: Eine oder mehrere Codezeilen, die zusammengehГ¶ren und ausgefГјhrt werden, wenn die Bedingung wahr ist
- **EinrГјckung**: In Python wird durch EinrГјckung (meist 4 Leerzeichen) gekennzeichnet, welche Zeilen zu einem Anweisungsblock gehГ¶ren

## 2. Die if-Anweisung: "Wenn... dann..." рџ›¤пёЏ

### Was ist eine if-Anweisung?

Die **if-Anweisung** ist die grundlegendste Form der Programmverzweigung in Python. Sie bedeutet wГ¶rtlich "**wenn**" und funktioniert nach einem einfachen Prinzip: **Wenn** eine Bedingung erfГјllt ist, **dann** wird der nachfolgende Code ausgefГјhrt. Wenn die Bedingung nicht erfГјllt ist, wird dieser Code einfach Гјbersprungen.

Denken Sie an eine automatische TГјr im Supermarkt: **Wenn** jemand sich der TГјr nГ¤hert (Bedingung erfГјllt), **dann** Г¶ffnet sich die TГјr (Aktion wird ausgefГјhrt). Wenn niemand da ist (Bedingung nicht erfГјllt), bleibt die TГјr geschlossen (Aktion wird nicht ausgefГјhrt).

### Wie schreibt man eine if-Anweisung?

Die Syntax (der Aufbau) einer if-Anweisung sieht in Python so aus:

```python
if Bedingung:
    # Anweisungsblock (wird nur ausgefГјhrt, wenn Bedingung True ist)
    # Diese Zeilen mГјssen eingerГјckt sein!
```

Wichtige Elemente:
1. **`if`**: Das SchlГјsselwort, das die Verzweigung einleitet
2. **Bedingung**: Ein Ausdruck, der zu `True` oder `False` ausgewertet wird
3. **Doppelpunkt `:`**: Markiert das Ende der Bedingung und den Beginn des Anweisungsblocks
4. **EinrГјckung**: Alle Zeilen, die zur if-Anweisung gehГ¶ren, mГјssen eingerГјckt sein (normalerweise 4 Leerzeichen)

### Wann wird der Code ausgefГјhrt?

Python prГјft die Bedingung nach dem `if`. Wenn diese Bedingung **wahr** (`True`) ist, fГјhrt Python alle eingerГјckten Zeilen aus. Wenn die Bedingung **falsch** (`False`) ist, Гјberspringt Python alle eingerГјckten Zeilen und macht mit dem Code nach der if-Anweisung weiter.

### Typische Einsatzgebiete

if-Anweisungen verwenden Sie Гјberall dort, wo Sie Code nur unter bestimmten UmstГ¤nden ausfГјhren mГ¶chten:
- Warnung ausgeben, wenn ein Grenzwert Гјberschritten wird
- Aktion ausfГјhren, wenn eine bestimmte Eingabe gemacht wurde
- Berechnungen nur fГјr positive Zahlen durchfГјhren
- Zugriff nur gewГ¤hren, wenn ein Passwort korrekt ist

### Beispiel 1: Einfache if-Anweisung mit positiver Zahl

````python
# Eine Variable mit einer positiven Zahl
x = 0

# PrГјfen, ob x grГ¶Гџer als 0 ist
if x > 0:
    print("x ist positiv")
    print("Der Wert von x ist:", x)

# Dieser Code wird immer ausgefГјhrt (nicht eingerГјckt)
print("Ende des Programms")
````

**ErklГ¤rung:**
- Die Variable `x` hat den Wert 10
- Die Bedingung `x > 0` wird geprГјft: Ist 10 grГ¶Гџer als 0? Ja! Also `True`
- Da die Bedingung wahr ist, werden die beiden eingerГјckten print-Anweisungen ausgefГјhrt
- Die letzte print-Anweisung wird immer ausgefГјhrt, da sie nicht eingerГјckt ist

### Beispiel 2: if-Anweisung mit negativer Zahl

````python
# Eine Variable mit einer negativen Zahl
x = -5

# PrГјfen, ob x grГ¶Гџer als 0 ist
if x > 0:
    print("x ist positiv")
    print("Der Wert von x ist:", x)

# Dieser Code wird immer ausgefГјhrt (nicht eingerГјckt)
print("Ende des Programms")
````

**ErklГ¤rung:**
- Die Variable `x` hat den Wert -5
- Die Bedingung `x > 0` wird geprГјft: Ist -5 grГ¶Гџer als 0? Nein! Also `False`
- Da die Bedingung falsch ist, werden die eingerГјckten print-Anweisungen **Гјbersprungen**
- Nur die letzte print-Anweisung wird ausgefГјhrt, da sie nicht eingerГјckt ist

### Beispiel 3: if-Anweisung mit Vergleich auf Gleichheit

````python
# Passwort-ГњberprГјfung (sehr vereinfacht!)
passwort = "geheim123"

if passwort == "geheim123" and True:
    print("Zugang gewГ¤hrt!")
    print("Willkommen im System.")
````

**ErklГ¤rung:**
- Die Variable `passwort` enthГ¤lt den String "geheim123"
- Mit dem Vergleichsoperator `==` prГјfen wir, ob der Inhalt von `passwort` genau "geheim123" ist
- Da dies der Fall ist (`True`), werden beide print-Anweisungen ausgefГјhrt

## 3. Die if-else-Anweisung: Entweder... oder... вљ–пёЏ

### Was ist eine if-else-Anweisung?

Oft reicht eine einfache if-Anweisung nicht aus. Manchmal mГ¶chten wir nicht nur etwas tun, **wenn** eine Bedingung erfГјllt ist, sondern auch etwas **anderes** tun, wenn sie **nicht** erfГјllt ist. Genau dafГјr gibt es die **if-else-Anweisung**.

Stellen Sie sich einen Lichtschalter vor: **Wenn** Sie ihn drГјcken und das Licht ist aus, **dann** geht es an. **Ansonsten** (also wenn das Licht schon an ist) geht es aus. Es gibt also immer zwei MГ¶glichkeiten вЂ“ entweder die eine Aktion oder die andere.

In der Programmierung bedeutet `else` genau das: "ansonsten" oder "in allen anderen FГ¤llen". Der Code im else-Block wird immer dann ausgefГјhrt, wenn die if-Bedingung **nicht** erfГјllt ist.

### Wie schreibt man eine if-else-Anweisung?

Die Syntax sieht so aus:

```python
if Bedingung:
    # Anweisungsblock 1 (wird ausgefГјhrt, wenn Bedingung True ist)
else:
    # Anweisungsblock 2 (wird ausgefГјhrt, wenn Bedingung False ist)
```

Wichtige Punkte:
- Das `else` steht auf derselben EinrГјckungsebene wie das `if`
- Nach `else` kommt auch ein Doppelpunkt `:`
- Der Code nach `else` muss ebenfalls eingerГјckt sein
- **Genau einer** der beiden BlГ¶cke wird ausgefГјhrt вЂ“ nie beide, nie keiner

### Warum ist if-else nГјtzlich?

Mit if-else kГ¶nnen Sie **garantieren**, dass Ihr Programm in jeder Situation eine Antwort hat. Es gibt keine "LГјcke" mehr вЂ“ entweder trifft die Bedingung zu, oder sie trifft nicht zu. In beiden FГ¤llen passiert etwas.

Typische AnwendungsfГ¤lle:
- Benutzereingabe prГјfen: Entweder korrekt oder inkorrekt
- Zugriff gewГ¤hren: Entweder Zugang oder Ablehnung
- Klassifizierung: Entweder positiv oder negativ/null
- Ampelsteuerung: Entweder grГјn (fahren) oder rot (stoppen)

### Der Unterschied zu einfachem if

Bei einem einfachen `if` kann es passieren, dass **kein** Code ausgefГјhrt wird (wenn die Bedingung falsch ist). Bei `if-else` wird **immer** einer der beiden BlГ¶cke ausgefГјhrt вЂ“ garantiert!

### Beispiel 1: Positive oder nicht-positive Zahl

````python
# Eine Variable mit einer negativen Zahl
x = 10

# PrГјfen und entsprechend reagieren
if x > 0:
    print("x ist positiv")
else:
    print("x ist nicht positiv (null oder negativ)")
````

**ErklГ¤rung:**
- Die Variable `x` hat den Wert -10
- Die Bedingung `x > 0` ist falsch (`False`), da -10 nicht grГ¶Гџer als 0 ist
- Daher wird der if-Block Гјbersprungen
- Stattdessen wird der else-Block ausgefГјhrt
- Ausgabe: "x ist nicht positiv (null oder negativ)"

### Beispiel 2: Produktionsziel erreicht oder nicht

````python
# Produktionsdaten
produzierte_teile = 85
tagesziel = 100

# ГњberprГјfung
if produzierte_teile >= tagesziel:
    print(f"Tagesziel erreicht! {produzierte_teile} Teile produziert.")
else:
    fehlende_teile = tagesziel - produzierte_teile
    print(f"Tagesziel nicht erreicht. Noch {fehlende_teile} Teile fehlen.")
````

**ErklГ¤rung:**
- Es wurden 85 Teile produziert, das Ziel liegt bei 100
- Die Bedingung `produzierte_teile >= tagesziel` ist falsch (85 ist nicht >= 100)
- Der else-Block wird ausgefГјhrt
- Im else-Block wird berechnet, wie viele Teile fehlen (100 - 85 = 15)
- Eine informative Nachricht wird ausgegeben

### Beispiel 3: VolljГ¤hrigkeit prГјfen

````python
# AltersГјberprГјfung
alter = 16

if alter >= 18:
    print("Zugang gewГ¤hrt вЂ“ Sie sind volljГ¤hrig.")
else:
    print("Zugang verweigert вЂ“ Sie mГјssen mindestens 18 Jahre alt sein.")
````

**ErklГ¤rung:**
- Das Alter ist 16 Jahre
- Die Bedingung `alter >= 18` ist falsch
- Der else-Block wird ausgefГјhrt und eine Ablehnungsmeldung ausgegeben

## 4. Die elif-Anweisung: Mehrere Bedingungen prГјfen рџ”Ђ

### Was ist eine elif-Anweisung?

Manchmal gibt es nicht nur zwei MГ¶glichkeiten (entweder-oder), sondern **mehrere verschiedene FГ¤lle**, die Sie unterscheiden mГ¶chten. Zum Beispiel: Eine Zahl kann positiv, negativ **oder** null sein вЂ“ das sind drei verschiedene FГ¤lle!

FГјr solche Situationen gibt es **elif** (eine AbkГјrzung fГјr "**else if**", also "ansonsten wenn"). Mit elif kГ¶nnen Sie mehrere Bedingungen nacheinander ГјberprГјfen, bis eine davon zutrifft.

Stellen Sie sich ein Ampelsystem vor: **Wenn** die Ampel grГјn ist, fahren Sie. **Ansonsten wenn** (elif) die Ampel gelb ist, werden Sie langsamer. **Ansonsten** (else) вЂ“ also bei rot вЂ“ halten Sie an. Drei verschiedene ZustГ¤nde, drei verschiedene Aktionen.

### Wie funktioniert elif?

Python geht die Bedingungen von oben nach unten durch:
1. Zuerst wird die if-Bedingung geprГјft
2. Wenn diese falsch ist, wird die erste elif-Bedingung geprГјft
3. Wenn auch diese falsch ist, wird die nГ¤chste elif-Bedingung geprГјft
4. Dieser Prozess geht so weiter, bis eine Bedingung wahr ist
5. **Sobald eine Bedingung wahr ist**, wird der zugehГ¶rige Block ausgefГјhrt und **alle weiteren Bedingungen werden Гјbersprungen**
6. Wenn keine Bedingung zutrifft und ein else vorhanden ist, wird der else-Block ausgefГјhrt

### Die Syntax

```python
if Bedingung1:
    # Code fГјr Fall 1
elif Bedingung2:
    # Code fГјr Fall 2
elif Bedingung3:
    # Code fГјr Fall 3
else:
    # Code fГјr alle anderen FГ¤lle
```

Wichtige Hinweise:
- Sie kГ¶nnen beliebig viele elif-Zweige haben (einen, zwei, zehn, ...)
- Das `else` am Ende ist optional вЂ“ Sie mГјssen es nicht verwenden
- Alle SchlГјsselwГ¶rter (if, elif, else) stehen auf derselben EinrГјckungsebene
- **Nur einer** der BlГ¶cke wird ausgefГјhrt, niemals mehrere!

### Warum ist elif besser als mehrere if?

Sie kГ¶nnten versucht sein, mehrere if-Anweisungen hintereinander zu schreiben. **Aber:** Bei mehreren if wГјrde Python **alle** Bedingungen prГјfen, auch wenn bereits eine erfГјllt war. Mit elif stoppt Python, sobald eine Bedingung zutrifft. Das ist effizienter und oft genau das, was Sie wollen.

### Typische AnwendungsfГ¤lle

- Notenberechnung: A, B, C, D, F je nach Punktzahl
- Altersgruppen: Kind, Jugendlicher, Erwachsener, Senior
- Produktionsauslastung: Niedrig, Normal, Hoch, Kritisch
- Temperaturklassifizierung: Kalt, KГјhl, Warm, HeiГџ

### Beispiel 1: Zahl klassifizieren (positiv, negativ, null)

````python
# Eine Zahl
x = 0

# Klassifizierung
if x > 0:
    print("x ist positiv")
elif x < 0:
    print("x ist negativ")
else:
    print("x ist null")
````

**ErklГ¤rung:**
- Die Variable `x` hat den Wert 0
- Erste PrГјfung: `x > 0`? Nein (0 ist nicht grГ¶Гџer als 0) в†’ if-Block wird Гјbersprungen
- Zweite PrГјfung: `x < 0`? Nein (0 ist nicht kleiner als 0) в†’ elif-Block wird Гјbersprungen
- Da beide Bedingungen falsch sind, wird der else-Block ausgefГјhrt
- Ausgabe: "x ist null"

### Beispiel 2: Produktionsauslastung bewerten

````python
# Produktionsdaten
produzierte_teile = 95
kapazitaet = 120  # Maximale TageskapazitГ¤t

# Auslastung berechnen (in Prozent)
auslastung = (produzierte_teile / kapazitaet) * 100

# Bewertung der Auslastung
if auslastung >= 90:
    print(f"Auslastung: {auslastung:.1f}% - Ausgezeichnet!")
elif auslastung >= 75:
    print(f"Auslastung: {auslastung:.1f}% - Gut")
elif auslastung >= 50:
    print(f"Auslastung: {auslastung:.1f}% - Befriedigend")
else:
    print(f"Auslastung: {auslastung:.1f}% - Zu niedrig!")
````

**ErklГ¤rung:**
- Es wurden 95 von 120 mГ¶glichen Teilen produziert
- Auslastung wird berechnet: (95/120) * 100 = 79,2%
- Erste Bedingung: `auslastung >= 90`? Nein (79,2 ist nicht >= 90)
- Zweite Bedingung: `auslastung >= 75`? Ja! (79,2 ist >= 75)
- Der elif-Block wird ausgefГјhrt, alle weiteren Bedingungen werden Гјbersprungen
- Ausgabe: "Auslastung: 79.2% - Gut"

### Beispiel 3: Altersklassifizierung

````python
# Alter einer Person
alter = 25

# Klassifizierung nach Altersgruppen
if alter < 12:
    print("Kind")
elif alter < 18:
    print("Jugendlicher")
elif alter < 65:
    print("Erwachsener")
else:
    print("Senior")
````

**ErklГ¤rung:**
- Das Alter ist 25 Jahre
- Erste Bedingung: `alter < 12`? Nein
- Zweite Bedingung: `alter < 18`? Nein
- Dritte Bedingung: `alter < 65`? Ja! (25 ist kleiner als 65)
- Der zweite elif-Block wird ausgefГјhrt
- Ausgabe: "Erwachsener"

## 6. Die match-case-Anweisung: Elegante Fallunterscheidung рџЋЇ

### Was ist match-case?

Ab Python 3.10 gibt es eine neue, elegante MГ¶glichkeit, verschiedene FГ¤lle zu unterscheiden: die **match-case-Anweisung**. Sie ist besonders nГјtzlich, wenn Sie eine Variable mit **vielen verschiedenen konkreten Werten** vergleichen mГ¶chten.

Stellen Sie sich ein MenГј in einem Restaurant vor: Je nachdem, welches Gericht Sie bestellen ("Pizza", "Pasta", "Salat"), passiert etwas anderes (verschiedene Zutaten werden vorbereitet). Mit match-case kГ¶nnen Sie solche FГ¤lle sehr Гјbersichtlich programmieren.

### Wann verwendet man match-case statt if-elif?

**match-case ist ideal, wenn:**
- Sie eine Variable mit vielen spezifischen Werten vergleichen ("A", "B", "C", "D")
- Sie exakte Гњbereinstimmungen prГјfen (Gleichheit)
- Sie viele FГ¤lle haben und der Code Гјbersichtlich bleiben soll

**if-elif ist besser, wenn:**
- Sie Bereiche prГјfen (z.B. `alter < 18`, `alter >= 65`)
- Sie komplexe Bedingungen haben (z.B. `alter > 18 and punkte >= 50`)
- Sie mit Г¤lteren Python-Versionen (< 3.10) arbeiten mГјssen

### Die Syntax von match-case

```python
match variable:
    case Wert1:
        # Code fГјr Wert1
    case Wert2:
        # Code fГјr Wert2
    case _:
        # Code fГјr alle anderen Werte (Standard-Fall)
```

Wichtige Elemente:
- **`match variable:`**: Gibt an, welche Variable geprГјft werden soll
- **`case Wert:`**: Definiert einen Fall вЂ“ wenn `variable` diesem Wert entspricht, wird der Code darunter ausgefГјhrt
- **`case _:`**: Der Unterstrich `_` ist ein Platzhalter fГјr "alle anderen FГ¤lle" (wie `else` bei if)
- Die case-BlГ¶cke mГјssen eingerГјckt sein

### Wie funktioniert match-case?

Python vergleicht den Wert der Variable nacheinander mit jedem case. Sobald eine Гњbereinstimmung gefunden wird, wird der zugehГ¶rige Code ausgefГјhrt und die match-Anweisung beendet. Wenn keine Гњbereinstimmung gefunden wird, wird der `case _:`-Block ausgefГјhrt (falls vorhanden).

### Vorteile von match-case

- **Гњbersichtlicher Code**: Besonders bei vielen FГ¤llen leichter zu lesen als lange if-elif-Ketten
- **Klarere Struktur**: Man sieht sofort, dass es um verschiedene FГ¤lle derselben Variable geht
- **Weniger Wiederholung**: Man muss die Variable nicht in jeder Bedingung neu schreiben

### Beispiel 1: Fruchtauswahl

````python
# Eine Variable mit einer Frucht
frucht = "Apfel"

# Match-Case-Anweisung
match frucht:
    case "Apfel":
        print("Ein Apfel hat etwa 95 Kalorien.")
    case "Banane":
        print("Eine Banane hat etwa 105 Kalorien.")
    case "Orange":
        print("Eine Orange hat etwa 62 Kalorien.")
    case _:
        print("Kalorienwert unbekannt.")
````

**ErklГ¤rung:**
- Die Variable `frucht` enthГ¤lt den String "Apfel"
- Python prГјft nacheinander: Ist `frucht` gleich "Apfel"? Ja!
- Der erste case-Block wird ausgefГјhrt
- Ausgabe: "Ein Apfel hat etwa 95 Kalorien."
- Alle weiteren cases werden Гјbersprungen

### Beispiel 2: Wochentage klassifizieren

````python
# Wochentag
tag = "Samstag"

# Klassifizierung
match tag:
    case "Montag":
        print("Wochenstart вЂ“ viel Energie!")
    case "Freitag":
        print("Fast Wochenende!")
    case "Samstag":
        print("Wochenende вЂ“ Zeit zum Entspannen!")
    case "Sonntag":
        print("Letzter freier Tag!")
    case _:
        print("Ein normaler Arbeitstag.")
````

**ErklГ¤rung:**
- Die Variable `tag` hat den Wert "Samstag"
- Python prГјft nacheinander die cases
- Beim dritten case gibt es eine Гњbereinstimmung: "Samstag" == "Samstag"
- Ausgabe: "Wochenende вЂ“ Zeit zum Entspannen!"

### Beispiel 3: Mehrere Werte in einem case (mit |)

````python
# Alter einer Person
alter = 2

# Altersklassifizierung mit match-case
match alter:
    case 0:
        print("Neugeborenes")
    case 1 | 2 | 3:
        print("Kleinkind")
    case 4 | 5 | 6:
        print("Vorschulkind")
    case _:
        print("Г„lteres Kind oder Erwachsener")
````

**ErklГ¤rung:**
- Die Variable `alter` hat den Wert 2
- Der erste case (0) passt nicht
- Der zweite case prГјft: Ist `alter` gleich 1 **ODER** 2 **ODER** 3?
- Ja! `alter` ist 2, also wird dieser Block ausgefГјhrt
- Der vertikale Strich `|` bedeutet "ODER" (eine von mehreren MГ¶glichkeiten)
- Ausgabe: "Kleinkind"

### Beispiel 4: match-case mit zusГ¤tzlichen Bedingungen (guards)

````python
# Alter mit zusГ¤tzlichen Bedingungen
alter = 15

match alter:
    case 0:
        print("Gerade geboren")
    case _ if alter < 12:
        print("Kind")
    case _ if alter < 18:
        print("Jugendlicher")
    case _ if alter < 65:
        print("Erwachsener")
    case _:
        print("Senior")
````

**ErklГ¤rung:**
- Hier kombinieren wir match-case mit Bedingungen (guards)
- `case _ if bedingung:` bedeutet: "Alle Werte, **wenn** die Bedingung erfГјllt ist"
- Das Alter ist 15
- Der erste case (0) passt nicht
- Der zweite case: `_ if alter < 12`? Nein (15 ist nicht < 12)
- Der dritte case: `_ if alter < 18`? Ja! (15 ist < 18)
- Ausgabe: "Jugendlicher"

**Hinweis:** In diesem Fall wГ¤re eine if-elif-Kette eigentlich Гјbersichtlicher. match-case mit guards ist eher fГјr spezielle Situationen gedacht.

## Zusammenfassung рџЋ“

### Was Sie in diesem Notebook gelernt haben:

- вњ… **if-Anweisungen**: Code nur unter bestimmten Bedingungen ausfГјhren
  - Syntax: `if Bedingung:` gefolgt von eingerГјcktem Code
  - Wird nur ausgefГјhrt, wenn die Bedingung `True` ist

- вњ… **if-else-Anweisungen**: Zwischen zwei Alternativen entscheiden
  - Entweder if-Block oder else-Block wird ausgefГјhrt (nie beide, nie keiner)
  - Garantiert, dass immer eine Aktion stattfindet

- вњ… **elif-Anweisungen**: Mehrere Bedingungen nacheinander prГјfen
  - Beliebig viele elif-BlГ¶cke zwischen if und else
  - Sobald eine Bedingung erfГјllt ist, werden alle weiteren Гјbersprungen
  - Effizienter als mehrere separate if-Anweisungen

- вњ… **match-case-Anweisungen**: Elegante Fallunterscheidungen
  - Ideal fГјr Vergleiche mit vielen spezifischen Werten
  - Mit `|` kГ¶nnen mehrere Werte in einem case kombiniert werden
  - `case _:` fungiert als Standard-Fall (wie else)

- вњ… **Bedingungen formulieren**: Mit Vergleichsoperatoren arbeiten
  - `>`, `<`, `>=`, `<=`, `==`, `!=` fГјr Vergleiche
  - Bedingungen ergeben immer `True` oder `False`

### Wichtige Konzepte:

| Konzept | Verwendung | Beispiel |
|---------|------------|----------|
| **if** | Eine Bedingung prГјfen | `if x > 0:` |
| **if-else** | Zwei Alternativen | `if x > 0: ... else: ...` |
| **elif** | Mehrere Bedingungen | `if ... elif ... else:` |
| **match-case** | Viele spezifische Werte | `match x: case "A": ...` |

### Wann verwende ich was?

- **Einfaches if**: Wenn Sie nur prГјfen mГ¶chten, ob etwas zutrifft (optional eine Aktion ausfГјhren)
- **if-else**: Wenn es zwei klare Alternativen gibt (entweder-oder)
- **if-elif-else**: Wenn Sie Bereiche oder komplexe Bedingungen prГјfen (z.B. Notenberechnung)
- **match-case**: Wenn Sie eine Variable mit vielen konkreten Werten vergleichen (z.B. Buchstaben, Wochentage)
